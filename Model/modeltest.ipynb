```python

#Import data into program

activity_training_data = pd.read_csv('train.csv')

# activity_test_data = pd.read_csv('test.csv')
```

```python

activity_sum = activity_training_data.groupby(['activity'])['rn'].sum()
activity_sum_df = activity_sum.to_frame()
x_train_coord = activity_training_data['activity'].unique()
y_train_coord = activity_sum_df['rn'].tolist()
#converting series to list
```

```python
# #plot box plots for independent variables
plt_1 = pyplot.figure(figsize=(15, 15))

pyplot.bar(x_train_coord, y_train_coord, color="red", width=0.7)
pyplot.show()

# pyplot.plot(f)
```

```python
# #Separate out dependent variable
y = activity_training_data.pop('activity')
x = activity_training_data.copy()
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#Split into test and train datasets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#Instantiate and fit model
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
#Predict for test set
predictions = model.predict(x_test)
```

```python
#Check accuracy of predictions
accuracy_score(y_test, predictions)
```

```python
#Plot confusion matrix

sklearn.metrics.plot_confusion_matrix(model, x, y)
```
